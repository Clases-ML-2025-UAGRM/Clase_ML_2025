```python
#Exactitud en la comprobación
knn.score(X_test, y_test)
```

### Análisis de los resultados

```python
import seaborn as sns

fig,ax = plt.subplots()
sns.scatterplot(data=producto1, hue='GRUPO_PRODUCTO', x='MESES_ANTIGUEDAD', y='INGRESO_NETO', palette="deep")
plt.legend(loc=2)
plt.savefig('scatter.png')
plt.show()
```

```python
knn.fit(X_train, y_train)
```

```python
#PRECISION VS TAMAÑO DE K EN la COMPROBACION
k_range = range(1,20)
scorese = []

for k in k_range:
    knn = KNeighborsClassifier(n_neighbors = k)
    scorese.append(knn.score(X_train, y_train))

plt.figure()
plt.xlabel('k')
plt.ylabel('exactitud')
plt.scatter(k_range, scorese, label='exactitud')
plt.legend()
plt.xticks([0,5,10,15,20]);
```

```python
#PRECISION VS TAMAÑO DE K EN la COMPROBACION
k_range = range(1,20)
scoresc = []

for k in k_range:
    knn = KNeighborsClassifier(n_neighbors = k)
    scoresc.append(knn.score(X_test, y_test))

plt.figure()
plt.xlabel('k')
plt.ylabel('exactitud')
plt.scatter(k_range, scoresc, label='exactitud')
plt.legend()
plt.xticks([0,5,10,15,20]);
```

```python
#PRECISION VS TAMAÑO DE K EN la COMPROBACION y EL ENTRENAMIENTO
k_range = range(1,20)
scorese = []

for k in k_range:
    knn = KNeighborsClassifier(n_neighbors = k)
    knn.fit(X_train, y_train)
    scorese.append(knn.score(X_train, y_train))


k_range = range(1,20)
scoresp = []

for k in k_range:
    knn = KNeighborsClassifier(n_neighbors = k)
    knn.fit(X_train, y_train)
    scoresp.append(knn.score(X_test, y_test))

plt.figure()
plt.xlabel('k')
plt.ylabel('exactitud')
plt.scatter(k_range, scoresp)
plt.scatter(k_range, scorese)
plt.legend(['test', 'train'])
plt.xticks([0,5,10,15,20]);
```

```python
#EXACTITUD VS TAMAÑO O PORPORCION DE MUESTRA DE ENTRENAMIENTO
t = [0.8, 0.6, 0.4, 0.2]

knn = KNeighborsClassifier(n_neighbors = 5)

plt.figure()

for s in t:
    scores = []
    for i in range(1,100):
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1-s)
        knn.fit(X_train, y_train)
        scores.append(knn.score(X_test, y_test))
    plt.plot(s, np.mean(scores), 'bo')

plt.xlabel('Training set proportion (%)')
plt.ylabel('accuracy');
```
